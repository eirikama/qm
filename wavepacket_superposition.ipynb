# Imports

```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib.animation import FuncAnimation

plt.style.use("default")
```

# Initialize Parameters and Define Functions

```python
hbar = 1.055 * 1e-34
m_el = 9.109 * 1e-31

L = 200 * 1e-9
sigma = 10 * 1e-9
x0 = 50 * 1e-9
E = 0.2*1.602 * 1e-19
```

```python
# function which returns the wavefunction of the wave packet
def psi_func(x, x0, sigma, E, m, t):    
    
    p = np.sqrt(2*m*E)
    omega = p**2/hbar / ( 2*m ) 
    
    C = np.sqrt(sigma) / ( np.pi**(0.25) * np.sqrt( sigma**2 + 1j*hbar*t / m ) )
    phase = np.exp( 1j*( (p/hbar)*(x-x0) - omega*t ))
    
    return C * phase * np.exp( -0.5*(x - x0 - (p/m)*t)**2 / ( sigma**2 + 1j*hbar*t / m ) ) 

def psi_n(n, x, L):
    return np.sqrt(2/L)*np.sin(n*np.pi*x/L)

def E_n(n, m, L):
    return n**2*np.pi**2*hbar**2/(2*m*L**2)

def plot_wavepacket():
    x = np.linspace(0, L, 1000)

    psi = psi_func(x, x0, sigma, t, E, m)
    psi_2 = np.abs(psi)**2
    
    plt.figure()
    plt.plot(x * 1e9, psi_2) 
    plt.xlabel(r"$x  (nm)$", fontsize=16)
    plt.ylabel(r"$|\Psi(x)|^2$", rotation=0, labelpad=25, fontsize=16) 
    plt.show()
```

```python
def trap(f, h):
    intgr = 0.5 * h * (f[0] + f[-1])
    for i in range(1, f.shape[0]-1):
        intgr = intgr + h * f[i]
    return intgr
```

# Wave Packet in Well

## Estimating the Coefficients 

```python
%matplotlib inline
```

```python
plt.style.use("default")
sigma = 20 * 1e-9
L = 200 * 1e-9
x0 = 50 * 1e-9

E = 0.2*1.602 * 1e-19 

N = 180
n_list = np.linspace(1, N, N)
x = np.linspace(0, L, 1000)
psi = psi_func(x, x0, sigma, E, m_el, t=0)
psi_2 = np.abs(psi) ** 2

cns = np.empty(shape=n_list.shape, dtype="complex128")
for n in n_list:
    cn = trap(np.conj(psi_n(n, x, L)) * psi, np.abs(x[1]-x[0]))
    cns[int(n-1)] = cn
    
P = np.zeros(shape=psi.shape, dtype="complex128")
norm = (np.abs(cns)**2).sum()
for n, cn in zip(n_list, cns):
    P += cn/norm * psi_n(n, x, L)
    
plt.plot(x * 1e9, np.abs(P)**2, label=f"$|\sum_{{n=1}}^{{N}} c_n \psi_n|^2$", color="indianred")
plt.plot(x * 1e9, psi_2, lw=3, label=f"$|\psi|^2$", ls="--", color="cornflowerblue")
plt.xlabel(r"$x (nm)$", fontsize=16)
plt.legend(fontsize=14)
plt.title(f"N={N}", fontsize=16)
plt.show()
```

```python
sigma = 20 * 1e-9
L = 200 * 1e-9
x0 = 50 * 1e-9

E = 0.2*1.602 * 1e-19 

N = 180
n_list = np.linspace(1, N, N)
x = np.linspace(0, L, 1000)
psi = psi_func(x, x0, sigma, E, m, t=0)
psi_2 = np.abs(psi) ** 2

cns = np.empty(shape=n_list.shape, dtype="complex128")
for n in n_list:
    cn = trap(np.conj(psi_n(n, x, L)) * psi, np.abs(x[1]-x[0]))
    cns[int(n-1)] = cn
    
P = np.zeros(shape=psi.shape, dtype="complex128")
norm = (np.abs(cns)**2).sum()
for n, cn in zip(n_list, cns):
    P += cn/norm * psi_n(n, x, L)

fig, (ax1, ax2)= plt.subplots(1, 2, figsize=(12,6))
ax1.plot(x * 1e9, psi_2, lw=3, label=f"$|\psi|^2$", ls="--", color="cornflowerblue")
ax2.plot(np.arange(1, cns.shape[0]+1), np.abs(cns), label=r"$|c_n|^2$", lw=3)
ax2.set_xlabel(r"$n$", fontsize=20)
ax1.set_xlabel(r"$x (nm)$", fontsize=20)
ax2.legend(fontsize=18)
ax1.legend(fontsize=18)
plt.show()
```

```python
sigma = 20 * 1e-9
L = 200 * 1e-9
x0 = 50 * 1e-9

E = 0.2*1.602 * 1e-19 

N = 500
n_list = np.linspace(1, N, N)
x = np.linspace(0, L, 1000)
psi = psi_func(x, x0, sigma, E, m, t=0)
psi_2 = np.abs(psi) ** 2

cns = np.empty(shape=n_list.shape, dtype="complex128")
for n in n_list:
    cn = trap(np.conj(psi_n(n, x, L)) * psi, np.abs(x[1]-x[0]))
    cns[int(n-1)] = cn

t_vals = np.linspace(0, 5000 * 1e-13, 400) 
norm = (np.abs(cns)**2).sum()

Pt = []
for t in t_vals:
    P = np.zeros(shape=psi.shape, dtype="complex128")
    for n, cn in zip(n_list, cns):
        time_prop = np.exp(-1j*E_n(n, m, L)*t/hbar)
        P += cn/norm * psi_n(n, x, L) * time_prop
    Pt.append(P)
```
